```python
import cv2
import os.path as osp
from matplotlib import pyplot as plt
from PageBlock import PageBlock
from Song import Song
from Key import Key
from Time import Time
from Signature import Signature
```

```python
# for keylet in ['C', 'Db', 'D', 'Eb', 'E', 'F', 'Gb', 'G', 'Ab', 'A', 'Bb', 'B', 'C#', 'D#', 'F#', 'G#', 'A#']:
for keylet in ['C', 'C#', 'D', 'Eb', 'E', 'F', 'F#', 'G', 'Ab', 'A', 'Bb', 'B']:
    for tonality in ['major', 'minor']:
        k = Key(keylet, tonality)
        t = Time(6, 8, 120, descriptor="allegro")
        s = Signature(k, t)
        song = Song("Autogenerated Song in %s %s" % (keylet, tonality), 'JT', s, original="Amadeus Chief Wolf")
        song.generate_chords(8)

        print("Autogenerated Song in %s %s" % (keylet, tonality))
        page = PageBlock()
        page.generate_blank_page()
        page.fill_with_song(song, write=False)
        page.save_minisheet('test/%s%s.jpg' % (keylet.replace('#', 'sharp'), tonality))
```
