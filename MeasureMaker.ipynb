```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
MEASURE_W = 300
MEASURE_H = 225
```

# First Measure of Line (Sharps)

```python
patch = np.zeros((MEASURE_H, MEASURE_W, 3), dtype=np.uint8) - 1

treble = cv2.imread("raw/treble.png")
bass   = cv2.imread("raw/bass.png")
leftbrace= cv2.imread("raw/staffbracket.png")

treble = cv2.cvtColor(cv2.resize(treble, (int(treble.shape[0] / 5), int(treble.shape[1] / 5))), cv2.COLOR_BGR2RGB)
treble = treble[:, 22:-22]

bass   = cv2.resize(bass, (int(bass.shape[0] / 12), int(bass.shape[1] / 9)))
leftbrace = cv2.resize(leftbrace, (int(leftbrace.shape[1] / 3), int(leftbrace.shape[0] / 1.9)))

patch[25:25+treble.shape[0], 20:20+treble.shape[1]] = treble
patch[150:150+bass.shape[0], 20:20+bass.shape[1]] = bass
patch[6:6+leftbrace.shape[0], 0:leftbrace.shape[1]] = leftbrace

#treble/bass maker
for i in range(5):
    cv2.line(patch, (20, 25 + 15*i), (300, 25 + 15*i), color =(0, 0, 0), thickness=1)
    cv2.line(patch, (20, 145+ 15*i), (300, 145 + 15*i), color =(0, 0, 0), thickness=1)
#bass maker

cv2.putText(patch, "#", (25 + treble.shape[1], 30), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # G
cv2.putText(patch, "#", (32 + treble.shape[1], 50), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # D
cv2.putText(patch, "#", (39 + treble.shape[1], 24), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # A
cv2.putText(patch, "#", (46 + treble.shape[1], 46), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # E
cv2.putText(patch, "#", (53 + treble.shape[1], 64), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # B
cv2.putText(patch, "#", (60 + treble.shape[1], 35), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # F#
cv2.putText(patch, "#", (67 + treble.shape[1], 59), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # C#

cv2.putText(patch, "#", (25 + treble.shape[1], 165), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # G
cv2.putText(patch, "#", (32 + treble.shape[1], 188), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # D
cv2.putText(patch, "#", (39 + treble.shape[1], 155), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # A
cv2.putText(patch, "#", (46 + treble.shape[1], 180), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # E
cv2.putText(patch, "#", (53 + treble.shape[1], 200), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # B
cv2.putText(patch, "#", (60 + treble.shape[1], 170), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # F#
cv2.putText(patch, "#", (67 + treble.shape[1], 190), fontScale=.75, 
            fontFace=cv2.FONT_HERSHEY_COMPLEX, color=(0, 0, 0), thickness=2) # C#





plt.imshow(patch)
plt.title("Measure")
plt.show()
```

# First Measure of Line (REPEAT)

```python
patch = np.zeros((MEASURE_H, MEASURE_W, 3), dtype=np.uint8) - 1

treble = cv2.imread("raw/treble.png")
bass   = cv2.imread("raw/bass.png")
leftbrace= cv2.imread("raw/staffbracket.png")

treble = cv2.cvtColor(cv2.resize(treble, (int(treble.shape[0] / 5), int(treble.shape[1] / 5))), cv2.COLOR_BGR2RGB)
treble = treble[:, 22:-22]

bass   = cv2.resize(bass, (int(bass.shape[0] / 12), int(bass.shape[1] / 9)))
leftbrace = cv2.resize(leftbrace, (int(leftbrace.shape[1] / 3), int(leftbrace.shape[0] / 1.9)))

patch[25:25+treble.shape[0], 20:20+treble.shape[1]] = treble
patch[150:150+bass.shape[0], 20:20+bass.shape[1]] = bass
patch[6:6+leftbrace.shape[0], 0:leftbrace.shape[1]] = leftbrace

#treble/bass maker
for i in range(5):
    cv2.line(patch, (20, 25 + 15*i), (300, 25 + 15*i), color =(0, 0, 0), thickness=1)
    cv2.line(patch, (20, 145+ 15*i), (300, 145 + 15*i), color =(0, 0, 0), thickness=1)

#repeat is thick, thin, colon
cv2.line(patch, (max(20+treble.shape[1], 25+bass.shape[1]), 25), (max(20+treble.shape[1], 25+bass.shape[1]), 205), 
         color =(0, 0, 0), thickness=2)
cv2.line(patch, (max(20+treble.shape[1], 25+bass.shape[1]) + 9, 25), (max(20+treble.shape[1], 25+bass.shape[1]) + 9, 205), 
         color =(0, 0, 0), thickness=1)
cv2.circle(patch, (max(20+treble.shape[1], 25+bass.shape[1]) + 16, 47), 3, (0, 0, 0), thickness=-1)
cv2.circle(patch, (max(20+treble.shape[1], 25+bass.shape[1]) + 16, 62), 3, (0, 0, 0), thickness=-1)
cv2.circle(patch, (max(20+treble.shape[1], 25+bass.shape[1]) + 16, 168), 3, (0, 0, 0), thickness=-1)
cv2.circle(patch, (max(20+treble.shape[1], 25+bass.shape[1]) + 16, 184), 3, (0, 0, 0), thickness=-1)


plt.imshow(patch)
plt.title("Measure")
plt.show()
```
