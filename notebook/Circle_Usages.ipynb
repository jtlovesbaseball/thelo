```python
circle_traversal = ['C', 'Db', 'D', 'Eb', 'E', 'F', 'Gb', 'G', 'Ab', 'A', 'Bb', 'B', 'C#', 'D#', 'F#', 'G#', 'A#']
for note in circle_traversal:
    circ = Circle(note)
    print ("Key: %s" % note)
    print ("Major: ", circ.generate_triad())
    print ("Minor: ", circ.generate_triad('minor'))
    print ("Dimin: ", circ.generate_triad('dim'))
    print ("Augme: ", circ.generate_triad('aug'))
    for st in ['7', 'maj7', 'min7', 'halfdim7', 'dim7']:
        print ('%8s: ' % st, circ.generate_extended(quality=st))
    print('---')
```
